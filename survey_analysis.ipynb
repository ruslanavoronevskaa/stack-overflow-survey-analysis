**Завдання 7. Географічний аналіз компенсації Python-розробників**.

**Опис завдання:** Серед респондентів, що програмують на Python, проведемо аналіз компенсації (ConvertedCompYearly) в розрізі країн.

**Для кожної країни обчислюємо:**

Середню суму компенсації


Медіанну суму компенсації


```python
# Відфільтровуємо респондентів, які працюють з Python
python_df = df1.loc[
    df1['LanguageHaveWorkedWith'].str.contains(
        'Python',
        case=False,
        na=False)]

#Видаляємо рядки з пропусками у компенсації
python_df = python_df.dropna(subset=['ConvertedCompYearly'])

# Групуємо по країнах та рахуємо середню і медіанну компенсацію
compensation_by_country = python_df.groupby('Country')['ConvertedCompYearly'].agg(
    average_compensation='mean',
    median_compensation='median'
).round(2).reset_index()

# Виводимо фінальний DataFrame
compensation_by_country
```

**Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів**

**Опис завдання:** Визначаємо рівні освіти 5 респондентів з найбільшою компенсацією. 

Потрібно:

Відсортувати респондентів за рівнем компенсації

Вибрати топ-5

Показати їх рівні освіти

```python
# Видаляємо рядки з пропусками у компенсації
df_compensation = df1.dropna(subset=['ConvertedCompYearly'])

# Сортуємо респондентів за компенсацією у спадаючому порядку
top_5_compensation = df_compensation.sort_values(
    by='ConvertedCompYearly',
    ascending=False)

# Вибираємо топ-5 респондентів
top_5_compensation = top_5_compensation.head(5)

# Виводимо рівні освіти та компенсацію
top_5_compensation = top_5_compensation[['ResponseId', 'ConvertedCompYearly', 'EdLevel']].reset_index(drop=True)
top_5_compensation
```

**Завдання 9. Аналіз популярності Python по віковим категоріям**


**Опис завдання:** Для кожної вікової категорії визначимо відсоток респондентів, які програмують на Python. Це дасть уявлення про те, в яких вікових групах Python є найпопулярнішим.

```python
# Всі респонденти по вікових категоріях
total_by_age = df1.groupby('Age')['ResponseId'].nunique()

# Респонденти, які програмують на Python
python_by_age = df1.loc[df1['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)].groupby('Age')['ResponseId'].nunique()

# Об'єднуємо в один DataFrame
python_share_by_age = (python_by_age / total_by_age * 100).round(2).reset_index(name='Python_percentage')

# Виводимо результат
python_share_by_age
```

**Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників**

**Опис завдання:** Знайдемо респондентів, які:

Знаходяться у 75 перцентилі за компенсацією (тобто заробляють більше, ніж 75% респондентів)

Працюють віддалено

Для цієї групи визначаємо найрозповсюдженіші індустрії.

```python
#Знаходимо значення  75 перцентилю 
q75 = df1['ConvertedCompYearly'].quantile(0.75)

#Знаходимо респондентів, що  заробляють більше, ніж 75% опитаних та працюють віддалено
resp_in_75_perc_remote = df1[(df1['ConvertedCompYearly'] > q75) & (df1['RemoteWork'] == 'Remote')]

#Визначаємо найрозповсюдженіші Industry
top_industries = resp_in_75_perc_remote['Industry'].value_counts().reset_index()

top_industries
```

```python
#Побудуємо візуалізацію з найпопулярнішими індустріями серед опитаних з найвищем рівнем заробітку та віддаленим форматом роботи
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.bar(
    top_industries['Industry'],
    top_industries['count']
)

plt.title('Top Industries among High-Earning Remote Users')
plt.xlabel('Industry')
plt.ylabel('Number of Respondents')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
